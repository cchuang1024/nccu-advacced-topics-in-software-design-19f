## 軟體設計

> Software design is the activity performed by a software developer that results in the software architecture of a system. It is concerned with specifying the component of a software system and the relationships between them within given functional and non-functional properties.

> 軟體設計開發人員主導的活動，其產出就是系統的架構設計。而架構的關鍵在基於功能性的，及非功能性的特性，來設計軟體系統的元件以及元件之間的關係。

**POSA vol 1**

* 軟體架構：軟體設計的產物

    - 非功能性的：關乎能做多好

## 軟體架構的定義

> Architecture is defined by the recommanded practice as the fundamental organization of a system, embodied in its components, their relationships to each other and the environment, and the principles governing its design and evolution.

**IEEE 1471-2000**

> The software architecture of a system is the set of structures needed to reason about the system, which comprise software elements, relations among them, and properties of both.

**Len Bass**

## design decisions

要滿足哪些系統品質，要做的取捨

1. 責任的分配
    - FR (functional requirement) or NFR (non-functional requirement)
1. 元件間的通訊方式
    - coordinate 屬性
        - timeliness, currency, completeness, consistency
    - coordinate 機制
        - 例：async/sync, stateful/stateless, ack/non-acked
1. 資料模型
    - key abstractions
    - 要做到大家都同意，就會很抽象而不切實際
    - DDD 提倡 Bounded context，只劃分小區域進行定義
    - CRC + Key Abstraction Form -> Data Model (簡式 class diagram)
1. 資源管理方式
    - 決定哪些資源要管控、誰來管控
    - 決定競爭資源時的仲裁機制
    - 資源不足時如何處理
1. 架構元素的 mapping，如 module 要算為 thread 或 process？
1. binding time decisions
    - 決定做決定的時機
1. choice of technology
    - 特定軟體技術的採用與衡量
    - IDE, 平台, 顧問...

### 三種架構的陳述型態

1. 靜態結構
